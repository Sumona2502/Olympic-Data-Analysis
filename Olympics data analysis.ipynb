### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df['Sport'].unique()
```

```python
df[df['Medal'] == 'Gold']
```

```python
df = df[df['Medal'] == 'Gold']
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport'] == Sport])])
    
pd.DataFrame(data , columns = ['Sport', 'freq']).sort_values(by = 'freq' , ascending = False).plot(x = 'Sport' , y = 'freq' , kind = 'bar' , figsize = (10,5))
```

Answer : Aquatics

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df[df['Medal'] == 'Gold']
```
